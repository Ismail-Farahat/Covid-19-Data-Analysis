```python
# To ignore unnecessary warrings
import warnings; warnings.simplefilter('ignore')

# To import the packages that will be used.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sps
import scipy.stats
import seaborn as sns

# To make figures show up
%matplotlib inline
```

```python
# Load the data 
df_china = pd.read_csv('COVID-19-daily-data-2020-04-12_china.csv')
df_usa = pd.read_csv('COVID-19-daily-data-2020-04-12_USA.csv')
df_italy = pd.read_csv('COVID-19-daily-data-2020-04-12_Italy.csv')
df_spain = pd.read_csv('COVID-19-daily-data-2020-04-12_spain.csv')

df_world = pd.read_csv('COVID-19-countries-data-2020-04-04_total.csv')
```

## China Data

```python
# To drop useless columns
df_china = df_china[['cases', 'moving_avg_temp', 'moving_avg_humidity', 'moving_avg_wind_speed', 'quarantine_days']]

# To drop useless rows
df_china = df_china.dropna()
```

```python
# To make the plot much bigger to make it clealy visible
plt.figure(figsize=(10,7))

# To build the matrix of correlation coefficient for all variables
corr_china = df_china.corr()

# To keep only the half of the matrix
mask = np.zeros_like(corr_china)
mask[np.triu_indices_from(mask)] = True

# To visualize the matrix
ax = sns.heatmap(
    corr_china, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(240, 10, n=9),
    annot=True,
    mask = mask,
    square=True)

# To set the labels of the axies
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right')
ax.set_yticklabels(
    ax.get_yticklabels(),
    rotation=45,
    horizontalalignment='right')

# To put a title on the polt
plt.title('FIG 1: Correlation Coefficients Matrix of China Data', {'fontsize': 14, 'fontweight' : 5});
```

```python
# To get the correlation coefficient of every factor
for index, row in corr_china.iterrows():
     display(row.to_frame())
     break
```

```python
# To make plot much bigger to make it clealy visible
plt.figure(figsize=(10,7))

# To plot the scatter plot and regresion model of linear relation between cases and qarantine days number
sns.regplot(x = df_china.quarantine_days,
            y = df_china.cases,
            scatter = True,
            fit_reg = True,
            color="b",
            ci = 80,
            order = 4)

# To set limits to axis
plt.ylim(0,5000)
plt.xlim(0,50)

# To put a title on the polt
plt.title('FIG 2: Number of Cases with Quarantine Days Number in China', {'fontsize': 14, 'fontweight' : 5})
plt.ylabel('Number of Cases', {'fontsize': 12, 'fontweight' : 5})
plt.xlabel('Quarantine Days Number', {'fontsize': 12, 'fontweight' : 5});
```
